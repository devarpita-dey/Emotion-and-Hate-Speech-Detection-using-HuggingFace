# Emotion Detection 

```python
emotion=create_analyzer(task='emotion', lang='en')
result=emotion.predict('I love to be an Data Scientist ')
result.output
```

```python
def emotion_detection(text):
    emotion=create_analyzer(task='emotion', lang='en')
    result=emotion.predict('I love to be an Data Scientist ')
    return result.output
```

```python
emotion_detection("I am happy")
```

# Hate Speech Detection

```python
hatespeech=create_analyzer(task='hate_speech', lang='en')
hatespeech.predict('I love playing cricket')
```

```python
hatespeech= create_analyzer(task='hate_speech', lang='en')
result=hatespeech.predict("I hate apples so I don't want to sit with an apple seller")
next(iter(result.probas.items()), None)
```

```python
def hatespeech_detection(text):
    hatespeech= create_analyzer(task='hate_speech', lang='en')
    result=hatespeech.predict(text)
    return next(iter(result.probas.items()), None)
hatespeech_detection("I hate apples so I don't want to sit with an apple seller")
```
